```python
import pandas as pd

# 1. Load the files
df_train = pd.read_csv('churn-bigml-80.csv')
df_test = pd.read_csv('churn-bigml-20.csv')

print("Files loaded successfully!")

# 2. Data Cleaning: Convert text/booleans to numbers (1 and 0)
# Convert 'Yes'/'No' to 1/0
df_train['International plan'] = df_train['International plan'].map({'Yes': 1, 'No': 0})
df_train['Voice mail plan'] = df_train['Voice mail plan'].map({'Yes': 1, 'No': 0})

df_test['International plan'] = df_test['International plan'].map({'Yes': 1, 'No': 0})
df_test['Voice mail plan'] = df_test['Voice mail plan'].map({'Yes': 1, 'No': 0})

# Convert Churn True/False to 1/0
df_train['Churn'] = df_train['Churn'].astype(int)
# df_test['Churn'] = df_test['Churn'].astype(int)

print("\n--- CLEANED DATA PEEK ---")
print(df_train[['International plan', 'Voice mail plan', 'Churn']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the look of our charts
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 5))

# CHART 1: Does complaining to Customer Service lead to higher Churn?
plt.subplot(1, 2, 1)
sns.barplot(data=df_train, x='Customer service calls', y='Churn', ci=None, palette='magma')
plt.title('Churn Rate vs Customer Service Calls')
plt.ylabel('Percentage of Customers who Left')
plt.xlabel('Number of Service Calls')

# CHART 2: Does having an International Plan cause higher Churn?
plt.subplot(1, 2, 2)
sns.countplot(data=df_train, x='International plan', hue='Churn', palette='Set1')
plt.title('Churn Count by International Plan')
plt.xlabel('Has International Plan? (0 = No, 1 = Yes)')
plt.ylabel('Number of Customers')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Select the columns we want to use to predict churn (dropping text/unneeded columns)
features = ['Account length', 'Area code', 'International plan', 'Voice mail plan',
            'Number vmail messages', 'Total day minutes', 'Total day calls', 'Total day charge',
            'Total eve minutes', 'Total eve calls', 'Total eve charge', 'Total night minutes',
            'Total night calls', 'Total night charge', 'Total intl minutes', 'Total intl calls',
            'Total intl charge', 'Customer service calls']

# Split training data
X_train = df_train[features]
y_train = df_train['Churn']

# Split testing data
X_test = df_test[features]
y_test = df_test['Churn']

print("Data successfully split and prepared for Machine Learning!")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Initialize the Decision Tree Model
model = DecisionTreeClassifier(max_depth=4, random_state=42)
# Note: max_depth=4 keeps the flowchart simple so it doesn't over-complicate things!

# 2. Train the Model (Teaching it using our training data)
model.fit(X_train, y_train)
print("AI Model Trained Successfully!")

# 3. Predict the Future (Testing it on our unseen test data)
y_pred = model.predict(X_test)

# 4. Check the Results!
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy Score: {accuracy * 100:.2f}%")
print("\nDetailed Performance Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd

# Show which features impacted the model the most
importance = pd.Series(model.feature_importances_, index=features)
importance.nlargest(3).plot(kind='barh', color='teal')
plt.title('Top 3 Predictive Features for Customer Churn')
plt.xlabel('Importance Score')
plt.show()
```
